```python
import yfinance as yf
import altair as alt
```

```python
data = yf.download('AAPL','2015-01-01','2019-12-18')
```

```python
data['date'] = data.index
```

```python
price = alt.Chart(data).mark_line().encode(
    x='date:T',
    y='Close',
    tooltip=['date', 'Open', 'Close', 'High', 'Low', 'Volume']
).properties(
    width=600, height=200
)

volume = alt.Chart(data).mark_line().encode(
    x='yearmonth(date):T',
    y='mean(Volume)').properties(width=600, height=60)

price & volume
```
